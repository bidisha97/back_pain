## Converting object to int of target variable...

```python
df['Class_att']=df['Class_att'].map({'Abnormal':1,
                                    'Normal':0})
```

```python
df.dtypes
```

```python
df.describe()
```

## Normalising the variables with StandardScaler...

```python
from sklearn.preprocessing import StandardScaler
```

## Segragating independent and dependent variable...

```python
x=df.drop(['Class_att'],axis=1)
y=df['Class_att']
```

```python
scaler=StandardScaler()
scaled_df=scaler.fit_transform(x)
scaled_df=pd.DataFrame(scaled_df,columns=['Col1','Col2','Col3','Col4','Col5','Col6','Col7','Col8','Col9','Col10','Col11','Col12'])
scaled_df.head()
```

```python
scaled_df.describe()
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.4,random_state=46,stratify=y)
```

## Implementing Logistic Regression...

```python
from sklearn.linear_model import LogisticRegression as LogReg
from sklearn.metrics import f1_score
```

```python
logreg=LogReg()
logreg.fit(train_x,train_y)
```

```python
train_predict=logreg.predict(train_x)
train_predict
```

```python
k=f1_score(train_predict,train_y)
print('Training f1_score',  k)
```

```python
test_predict=logreg.predict(test_x)
k=f1_score(test_predict,test_y)
print('Test f1_score',  k)
```

```python
train_predict=logreg.predict_proba(train_x)
train_predict
```

```python
train_preds=train_predict[:,1]
train_preds
```

```python
for i in range(0, len(train_preds)):
  if(train_preds[i]>0.55):
    train_preds[i] = 1
  else:
    train_preds[i] = 0
  
```

```python
k=f1_score(train_preds,train_y)
print('Training F1_score',  k)
```

```python
from sklearn.metrics import confusion_matrix
cf= confusion_matrix(test_y, test_predict)
print(cf)
```

```python
from sklearn.metrics import classification_report as rep
print(rep( test_y , test_predict ))
```

```python
logreg.coef_
```

```python
plt.figure(figsize=(8, 6), dpi=120, facecolor='w', edgecolor='b')
x = range(len(train_x.columns))
c = logreg.coef_.reshape(-1)
plt.bar( x, c )
plt.xlabel( "Variables")
plt.ylabel('Coefficients')
plt.title('Coefficient plot')
```

```python
Coefficients = pd.DataFrame({
    'Variable'    : train_x.columns,
    'coefficient' : abs(c)
})
Coefficients.head(13)
```
